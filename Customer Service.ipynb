```python
missing_value_percentage=(data.isna().sum(axis=0)/data.shape[0])*100
```

```python
missing_value_percentage
```

```python
def drop_columns(data,threshold):
    missing_percentage=(data.isna().sum()/len(data))*100
    columns_to_drop=missing_percentage[missing_percentage>threshold].index
    data.drop(columns=columns_to_drop,inplace=True)
    return data
threshold=80
```

```python
data=drop_columns(data,threshold)
```

```python
data.shape
```

```python
(data.isna().sum()/data.shape[0]*100).sort_values(ascending=False)[:10]
```

```python
data=data.dropna(axis=0,how='any')
```

```python
data.isna().sum()
```

```python
from datetime import date,time,datetime
data['Created Date']=data['Created Date'].astype('datetime64[ns]')
data['Closed Date']=data['Closed Date'].astype('datetime64[ns]')
data['Due Date']=data['Due Date'].astype('datetime64[ns]')
```

```python
data['Time elapsed']=data['Closed Date']-data['Created Date']
```

```python
data['Resolution_time']=(data['Time elapsed'].astype('timedelta64[ns]'))//(60*60)
```

```python
crosstabresult=pd.crosstab(index=data['Complaint Type'],columns=data['Location'])
from scipy.stats import chi2_contingency
CHISqresult=chi2_contingency(crosstabresult)
print('The P-value of the CHISq Test is:',CHISqresult[1])
```

```python
#the second value of a statistical test is alwaysthe p-value
#as the p-value is less than 0.5,we will reject the null hypothesis
#H0:there is no relationship between complaint_type and location
#H1:there is relationship
```
